```python
Concrete_df.drop_duplicates( keep='first', inplace=False)
```

```python
Concrete_df.columns
```

```python
Concrete_df=Concrete_df.rename(columns={
    'Cement (component 1)(kg in a m^3 mixture)':'Cement (component 1)' ,
       'Blast Furnace Slag (component 2)(kg in a m^3 mixture)':'Blast',
       'Fly Ash (component 3)(kg in a m^3 mixture)':'FlyAsh',
       'Water  (component 4)(kg in a m^3 mixture)':'Water',
       'Superplasticizer (component 5)(kg in a m^3 mixture)':'Superplasticizer',
       'Coarse Aggregate  (component 6)(kg in a m^3 mixture)':'Coarse Aggregate ',
       'Fine Aggregate (component 7)(kg in a m^3 mixture) Age (day)': 'Fine Aggregate',
        'Concrete compressive strength(MPa, megapascals) ':'ConcreteTarget',

})
Concrete_df
```

```python
flyash=Concrete_df['FlyAsh'].mean()
flyash
```

```python
BlastFurnaceSlag=Concrete_df['Blast'].mean()
BlastFurnaceSlag
```

```python
BlastFurnaceSlag=Concrete_df['Blast'].mean()
```

```python
Concrete_df.Blast=Concrete_df.Blast.mask(Concrete_df.Blast==0).fillna(BlastFurnaceSlag)
Concrete_df
Concrete_df.FlyAsh=Concrete_df.FlyAsh.mask(Concrete_df.FlyAsh==0).fillna(flyash)
Concrete_df
```

```python
y = Concrete_df.ConcreteTarget
x=Concrete_df.drop(columns='ConcreteTarget')
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
print(X_train.shape)
X_test.shape
```

```python
X_train, X_val, Y_train, Y_val = train_test_split(X_train, Y_train, test_size=0.2)
```

```python
mean = X_train.mean(axis=0)
std = X_train.std(axis=0)

X_train -= mean
X_train /= std

X_test -= mean
X_test /= std


X_val -= mean
X_val /= std
```

```python
X_train.head()
```

```python

from tensorflow.keras.models import Sequential
from tensorflow.keras import layers
```

```python
network = Sequential()
network.add(layers.Dense(6, activation='relu', input_shape=(X_train.shape[1],)))
network.add(layers.Dense(4, activation='relu'))
network.add(layers.Dense(1))
```

```python
network.compile(optimizer='adam', loss='mse', metrics='mae')
```

```python
history = network.fit(X_train, Y_train, batch_size=30, epochs=200, validation_data=(X_val, Y_val))
```

```python
epochs=range(0,200)
train_loss = history.history['loss']
val_loss = history.history['val_loss']

plt.plot(epochs, train_loss, 'r', label='Training Loss')
plt.plot(epochs, val_loss, 'g', label='Validation Loss')
plt.title('Training and Validation loss ')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
